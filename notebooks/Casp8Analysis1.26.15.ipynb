```python
def df_to_names(df):
    
    name_dict = {}
        
    for i in range(1,len(df.columns)):
        name = df.columns[i][:-2]
        if name in name_dict:
            name_dict[name].append(i)
        else:
            name_dict[name] = [i]
            
    return name_dict
```
